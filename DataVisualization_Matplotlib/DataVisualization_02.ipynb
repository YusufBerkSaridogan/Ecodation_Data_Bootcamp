```python
plt.figure(figsize = (6,6))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45)
plt.show()
```

```python
plt.figure(figsize = (8,8))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#7CB342', '#C0CA33', '#FFB300', '#F57C00']
plt.pie(area, labels = labels, colors = colors, startangle = 45, shadow = 'true', autopct = '%1.2f%%')
plt.show()
```

```python
plt.figure(figsize = (8,8))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#7CB342', '#C0CA33', '#FFB300', '#F57C00']
def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x, sum(area)*x/100)
plt.pie(area, labels = labels, colors = colors, startangle = 45, shadow = 'true', autopct = val_per)
plt.show()
```

```python
#Donut Plot
```

```python
plt.figure(figsize = (8,8))
area = [60, 35, 10]
labels = ['Sold Tickets', 'Unsold Tickets', 'Cancelled Tickets']
colors = ['green', 'blue', 'red'] 
plt.pie(area, labels= labels, colors= colors, startangle= 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf() #get current figure
p.gca().add_artist(my_circle) #get current axes
plt.show()
```

```python
fig = plt.figure(figsize = (8,8))
fig.patch.set_facecolor('#DADADA')
area = [60, 35, 10]
labels = ['Sold Tickets', 'Unsold Tickets', 'Cancelled Tickets']
colors = ['green', 'blue', 'red'] 
plt.pie(area, labels= labels, colors= colors, startangle= 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf() #get current figure
p.gca().add_artist(my_circle) #get current axes
plt.show()
```

```python
fig = plt.figure(figsize = (8,8))
fig.patch.set_facecolor('#DADADA')
area = [60, 35, 10]
labels = ['Sold Tickets', 'Unsold Tickets', 'Cancelled Tickets']
colors = ['green', 'blue', 'red'] 
plt.pie(area, labels= labels, colors= colors, startangle= 45, autopct= '%1.2f%%',explode = [0, 0.03, 0.03,])
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf() #get current figure
p.gca().add_artist(my_circle) #get current axes
plt.show()
```

```python
fig = plt.figure(figsize = (8,8))
fig.patch.set_facecolor('#DADADA')
area = [60, 35, 10]
labels = ['Sold Tickets', 'Unsold Tickets', 'Cancelled Tickets']
colors = ['green', 'blue', 'red'] 
plt.pie(area, labels= labels, colors= colors, startangle= 45, autopct= '%1.2f%%', pctdistance= 0.85, explode = [0, 0.03, 0.03,])
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf() #get current figure
p.gca().add_artist(my_circle) #get current axes
plt.show()
```

```python
fig = plt.figure(figsize = (8,8))
fig.patch.set_facecolor('#DADADA')
area = [60, 35, 10]
labels = ['Sold Tickets', 'Unsold Tickets', 'Cancelled Tickets']
colors = ['green', 'blue', 'red'] 
def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x, np.sum(area)*x/100)
plt.pie(area, labels= labels, colors= colors, startangle= 45, autopct= val_per, pctdistance= 0.85, explode = [0, 0.03, 0.03,])
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf() #get current figure
p.gca().add_artist(my_circle) #get current axes
plt.show()
```

```python
#Multiple Donut Pie Charts
```

```python
fig = plt.figure(figsize = (20, 6))
area = [48, 30, 20, 15]
priority = ['Low', 'Medium', 'High', 'Critical']
status = ['Resolved', 'Cancelled', 'Pending', 'Assigned']
company = ['Tesla', 'Microsoft', 'Apple', 'TOGG']
colors = ['#7BC342', '#C0CA33', '#FFB300', '#F57C00']

plt.subplot(1,3,1)
plt.pie(area, labels = priority, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.subplot(1,3,2)
plt.pie(area, labels = status, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.subplot(1,3,3)
plt.pie(area, labels = company, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
fig = plt.figure(figsize = (20, 6))
area = [48, 30, 20, 15]
priority = ['Low', 'Medium', 'High', 'Critical']
status = ['Resolved', 'Cancelled', 'Pending', 'Assigned']
company = ['Tesla', 'Microsoft', 'Apple', 'TOGG']
colors = ['#7BC342', '#C0CA33', '#FFB300', '#F57C00']

plt.subplot(3,1,1)
plt.pie(area, labels = priority, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.subplot(3,1,2)
plt.pie(area, labels = status, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.subplot(3,1,3)
plt.pie(area, labels = company, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
#Histogram; Distribution of a numerical variable
```

```python
x = np.random.normal(size = 3000)
plt.hist(x, bins = 40, color = 'brown', edgecolor= 'black')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
plt.figure(figsize = (12,6))
x = np.random.normal(size = 3000)
plt.hist(x, bins = 40, color = 'brown', edgecolor= 'black')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
#Multiple Histograms
```

```python
plt.figure(figsize = (8, 5))
x = np.random.normal(-4, 1, size = 800)
y = np.random.normal(0, 1.5, size = 800)
z = np.random.normal(3.5, 1, size = 800)
plt.hist(x, bins = 30, color = 'yellowgreen', alpha = 0.6, edgecolor= 'black')
plt.hist(y, bins = 30, color = '#FF8F00', alpha = 0.6,edgecolor= 'black')
plt.hist(z, bins = 30, color = 'blue', alpha = 0.6,edgecolor= 'black')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
#Using Histograms Cumulative Distribution Function (CDF) drawing
```
