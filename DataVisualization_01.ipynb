```python
plt.figure(figsize=(10,5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x,2)
y2 = np.power(x,3)
plt.plot(x, y1, 'b-',label= 'y1 = x^2')
plt.plot(x, y2, 'r-',label= 'y1 = x^3')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x,2)
y2 = np.power(x,3)
plt.plot(x, y1, 'b-',label= 'y1 = x^2')
plt.plot(x, y2, 'r-',label= 'y1 = x^3')
plt.xlabel('X Axis', fontsize=18)
plt.ylabel('Y Axis', fontsize=18)
plt.title('Graph', fontsize=20, color='purple')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
x = np.linspace(0,10,2000)

plt.plot(x, np.sin(x), 'b',label= 'Sinx')
plt.plot(x, np.cos(x), 'r',label= 'Cosx')
plt.xlabel('X Axis', fontsize=18)
plt.ylabel('Y Axis', fontsize=18)
plt.title('Sinx ve Cosx Graphs', fontsize=20, color='purple')
plt.tight_layout()
plt.legend(loc= 'lower left')
plt.show()
```

```python
plt.figure(figsize=(10,5))
x = np.linspace(0,10,2000)

plt.plot(x, np.sin(x), 'b',label= 'Sinx')
plt.plot(x, np.cos(x), 'r',label= 'Cosx')
plt.fill_between(x,0, np.sin(x))
plt.fill_between(x,0, np.cos(x))
plt.xlabel('X Axis', fontsize=18)
plt.ylabel('Y Axis', fontsize=18)
plt.title('Sinx ve Cosx Graphs', fontsize=20, color='purple')
plt.tight_layout()
plt.legend(loc= 'lower left')
plt.show()
```

```python
#Multiple Graphs
```

```python
plt.figure(figsize=(14,6))
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.show()
```

```python
plt.figure(figsize=(14,6))
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y1, 'r-')
plt.subplot(2,1,2)
plt.plot(x,y2, 'b-')
plt.tight_layout()
plt.show()
```

```python
#Bar Graphs 
```

```python
id1 = np.arange(1,10)
score = np.arange(20,110,10)
plt.bar(id1,score)
plt.xlabel('Student ID')
plt.ylabel('Score')
plt.show()
```

```python
id1 = np.arange(1,10)
score = np.arange(20,110,10)
plt.figure(figsize=(10,5))
ax = plt.axes()
ax.set_facecolor("white")
plt.bar(id1,score, color= '#FFA726')
plt.xlabel('Student ID')
plt.ylabel('Score')
plt.show()
```

```python
x1 = [1,3,5,7]
x2 = [2,4,6,8]
y1 = [7,7,7,7]
y2 = [17,18,31,50]
plt.figure(figsize=(10,5))
ax = plt.axes()
ax.set_facecolor("white")
plt.bar(x1,y1, label= 'First',color= 'red')
plt.bar(x2, y2, label='Second', color= 'blue')
plt.legend()
plt.title('Bar Graphs')
plt.show()
```

```python
Age = [24,15,89,57,21]
Name = ['Yusuf','Berk','Dou','Ege','Fatma']
plt.barh(Name,Age, color= 'purple')
plt.show()
```

```python
num1 = np.array([1,3,5,7,9])
num2 = np.array([2,4,6,8,10])
plt.figure(figsize=(12,5))
plt.bar(num1, num1**2, width = 0.3, color='orange')
plt.bar(num2, num2**2, width = 0.3, color='red')
for x,y in zip(num1,num1**2):
    plt.text(x, y + 0.05, '%d' % y, ha = 'center', va = 'bottom')
for x,y in zip(num2,num2**2):
    plt.text(x, y + 0.05, '%d' % y, ha = 'center', va = 'bottom')
plt.show()
```
