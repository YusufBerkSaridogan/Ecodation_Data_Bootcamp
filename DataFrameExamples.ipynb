```python
#DataFrame Merge
daf1 = pd.DataFrame({'id' : ['1','2','3','4','5'], 'Name' : ['Ahmet', 'Mehmet', 'Ayşe', 'Berk','Berivan']})
daf1
```

```python
daf2 = pd.DataFrame({'id' : ['1','2','6','7','8'], 'Score' : [40,55,90,21,79]})
daf2
```

```python
#Inner Join
pd.merge(daf1,daf2, on='id', how='inner')
```

```python
#Full Outer Join
pd.merge(daf1,daf2, on='id', how='outer')
```

```python
#Left Outer Join
pd.merge(daf1,daf2, on='id', how='left')
```

```python
#Right Outer Join
pd.merge(daf1,daf2, on='id', how='right')
```
